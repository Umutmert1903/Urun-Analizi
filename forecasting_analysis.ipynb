# Satış Tahmini (Forecasting)
Bu notebook, sezonluk satış verileri üzerinden 2025 Bahar tahmini üretir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Veri
data = {'Season': ['Spring', 'Summer', 'Fall', 'Winter'],
        'Total_Sales': [55200, 43100, 72000, 61000]}
df = pd.DataFrame(data)
df.index = pd.date_range(start='2024-03', periods=4, freq='Q')

# Model
x = np.arange(len(df)).reshape(-1, 1)
y = df['Total_Sales'].values
model = LinearRegression().fit(x, y)
future_index = np.array([[len(df)]])
predicted = model.predict(future_index)[0]

# Grafik
future_date = df.index[-1] + pd.DateOffset(months=3)
plt.plot(df.index, df['Total_Sales'], marker='o', label='Gerçek Satışlar')
plt.plot(future_date, predicted, marker='x', color='red', label='2025 Spring Tahmini')
plt.title('Sezona Göre Satış ve Tahmin')
plt.ylabel('Toplam Satış (USD)')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
print(f'2025 Spring sezonu için tahmin edilen satış: ${predicted:,.2f}')
```
